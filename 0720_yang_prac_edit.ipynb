```
def start_command():
    first_n = int(input('명령 횟수를 입력하세요 : ')) #first_n = 명령 횟수(첫 번째 입력값)
    # 1<=N<=10000 조건 맞추기
    if first_n > 10000:
      print("1이상 10,000이하의 정수 값을 입력해주세요.")
      first_n = start_command()

    elif first_n <= 0:
      print("1이상 10,000이하의 정수 값을 입력해주세요.")
      first_n = start_command()


    #N의 조건 충족할 경우;
    else:
      st_list = [] #st_list = 스택 저장소
      count = 0
      m = 0 #실제 명령 입력한 횟수('first_n' 이후 입력한 값의 $횟수$)

      while True:
        all_inp = input("push, size, pop, empty, top 중 명령 하나 입력하고, 필요시 space 1번 후 숫자를 입력해주세요 : ").split() #first_n 다음 입력창(반복할 입력창)
        str_inp, int_inp = all_inp # 명령어랑 숫자 구분
        int_inp = int(int_inp) #숫자 타입을 정수로 변환
        m += 1
        if m == first_n:
          break
        else:
          if str_inp == 'push':
            count += 1 #size 명령어 대비, 스택 저장소에 저장
            st_list.append(int_inp)

          elif str_inp == 'size':
            print(count)

          elif str_inp == 'pop':
            st_list.pop()
            count -= 1
            print(count)

          elif str_inp == 'empty':
            if st_list == 0 :
              print(1)
            else :
              print(0)

          elif str_inp == 'top':
            if st_list== 0:
              print(-1)
            else :
              print(st_list[-1]) # Q.이거 안 쓰고는 어떻게 하지?

          else :
            print("잘못 입력하셨습니다.제대로 된 명령어와 필요 시, 숫자를 입력해주세요.")


start_command()

# # while 무한 루프
# while True :
#     number = int(input("숫자를 입력하세요: "))
#     count += 1
#     if correct_answer == number : # 정답이면 break
#         break
#     elif correct_answer > number :
#         print(number, "보다 큽니다!")
#     else :
#         print(number, "보다 작습니다!")
```
